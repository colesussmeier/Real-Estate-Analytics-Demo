```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('ZHVI/ZHVI_December22.csv', dtype=str)
```

```python
df = df.loc[:,['RegionName', '2022-12-31']]
df
```

```python
num_months_updated = 1
arr = np.empty([3,num_months_updated * df['RegionName'].nunique()], dtype=object)
```

```python
a = 0
b = num_months_updated

# for every zip, transpose date columns and repeat zip code in 3rd column
for ind, row in df.iterrows():
    
  timeseries = row.iloc[1:].T

  arr[0,a:b] = timeseries.index.values
  arr[1,a:b] = timeseries.values
  arr[2,a:b] = np.repeat(row['RegionName'], len(timeseries.index.values))
  a += len(timeseries)# length of time in dataset
  b += len(timeseries)
```

```python
update_df = pd.DataFrame(arr.T)
update_df.columns = ['Date', 'ZHVI', 'ZipCode']
```

```python
update_df
```

```python
# exports in 3rd normal form
```

```python
update_df['ZHVI'] = update_df['ZHVI'].astype(float)
```

```python
update_df['ZHVI'] = update_df['ZHVI'].astype(int)
```

```python
update_df.to_csv('update_December22.csv', index= False)
```

```python
zips['Location'] = zips.apply(lambda x: sr.by_zipcode(x['Zip']).major_city + ', ' + sr.by_zipcode(x['Zip']).state + " " + str(x['Zip']), axis=1)
```
